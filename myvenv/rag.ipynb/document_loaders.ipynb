IN RAG- Data Ingestion is the most important part without data we cannot do anything 

various document_loaders in langchain_community

1.TextLoader is used to load the txt

```python
from langchain_community.document_loaders import TextLoader
```

```python
loader=TextLoader('sample.txt')
docs=loader.load()
```

```python
print(type(docs))
print(docs)
```

2.pypdfloader is used to load the pdf files 

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdfloader=PyPDFLoader('arxiv.pdf')
docs1=pdfloader.load()
```

```python
print(docs1)
```

#webbaseloader is used to load the webpage for the given url

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
webloader=WebBaseLoader(web_path="https://www.geeksforgeeks.org/introduction-deep-learning/")
docs2=webloader.load()
```

```python
print(docs2)
```

4.arxivloader is used to load arxiv page

```python
from langchain_community.document_loaders import ArxivLoader
```

```python
arxivloader=ArxivLoader(query='1706.03762',doc_content_chars_max=50)
docs4=arxivloader.load()
```

```python
print(docs4)
```

5.wikipedialoader is used to load the page content based on user request

```python
from langchain_community.document_loaders import WikipediaLoader
```

```python
wikiloader=WikipediaLoader(query='deep learning',load_max_docs=5)
doc5=wikiloader.load()
```

```python
print(doc5)
```
