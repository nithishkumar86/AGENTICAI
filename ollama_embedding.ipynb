```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
```

```python
pdfloader=PyPDFLoader('arxiv.pdf')
docs2=pdfloader.load()
```

```python
print(docs2)
```

```python
#RecursiveCharacterTextSplitter
splitter1=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
documents1=splitter1.split_documents(docs2) #documents will only be list
```

```python
documents1
```

```python
embedding=OllamaEmbeddings(model="mxbai-embed-large")
```

```python
db=FAISS.from_documents(documents=documents1,embedding=embedding)
```

```python
query="what is transformer in gen ai"
response=db._embed_query(query)
print(response)
```

#similarity seach in db

```python
result1=db.similarity_search(query)
print(result1)
```

#similarity search with score

```python
result2=db.similarity_search_with_relevance_scores(query)
```
