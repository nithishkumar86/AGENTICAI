various textsplitting techniques in langchain

textsplitter is used to chunk the data into small documents because all the  llm models have 
some limited context size sometimes we get an error we reach beyond the level 

1.Recursive_character_text_splitter-by default it can split like that ['\n\n,\n,' ','']

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader=TextLoader('sample.txt')
docs=loader.load()
```

```python
print(docs)
```

```python
#RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=20)
documents=splitter.split_documents(docs) #documents will only be list
```

```python
print(documents,end=" ")
```

```python
print(len(documents))
print(documents[0])
print(documents[1])
print(type(documents[0]))
```

2.character_text_splitter

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
```

```python
pdfloader=PyPDFLoader('arxiv.pdf')
docs2=pdfloader.load()
```

```python
print(docs2)
```

```python
#RecursiveCharacterTextSplitter
splitter1=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
documents1=splitter1.split_documents(docs2) #documents will only be list
```

```python
print(documents1,end=" ")
```

```python
print(len(documents1))
```

```python
print(documents1[0])
print(documents1[1])
```

#HTMLTextSplitter

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
from langchain_community.document_loaders import WebBaseLoader
```

```python
header_to_split_on=[
    ('h1','header1'),
    ('h2','header2'),
    ('h3','header3')
]
```

```python
htmlsplitter=HTMLHeaderTextSplitter(header_to_split_on)
documents2=htmlsplitter.split_text_from_url("https://www.geeksforgeeks.org/introduction-deep-learning/")
```

```python
paragraph="""
As we dive into this introductory exploration of Deep Learning,
 we uncover its foundational principles, applications, and the underlying mechanisms that
empower machines to achieve human-like cognitive abilities. This article serves as a gateway
into understanding how Deep Learning is reshaping industries, pushing the boundaries of
what’s possible in AI, and paving the way for a future where intelligent systems can
perceive, comprehend, and innovate autonomously.


The definition of Deep learning is that it is the branch of machine learning that is based 
on artificial neural network architecture. An artificial neural network or ANN uses layers of 
interconnected nodes called neurons that work together to process and learn from the input data.

In a fully connected Deep neural network, there is an input layer and 
one or more hidden layers connected one after the other. Each neuron receives input 
from the previous layer neurons or the input layer. The output of one neuron becomes the
input to other neurons in the next layer of the network, and this process continues until
the final layer produces the output of the network. The layers of the neural network
transform the input data through a series of nonlinear transformations,
      
 allowing the network to learn complex representations of the input data.
"""
```

```python
textsplitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=50)
documents4=textsplitter.create_documents([paragraph])
```
