#Text Embeddings: Hugging Face offers several models for text embeddings, such as the Xenova/

gte-small model. These models can be customized and run locally or accessed through APIs2.

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['HUGGINGFACE_API_KEY']=os.getenv('HUGGINGFACE_API_KEY')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
```

```python
embedding=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
pdfloader=PyPDFLoader('arxiv.pdf')
docs2=pdfloader.load()
```

```python
print(docs2)
```

```python
#RecursiveCharacterTextSplitter
splitter1=CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
documents1=splitter1.split_documents(docs2) #documents will only be list
```

```python
documents1
```

```python
db=FAISS.from_documents(documents=documents1,embedding=embedding)
```

```python
#storing fiass db
db.save_local('faiss-vectorstore')
```

```python
import pickle
```

```python
load_db=FAISS.load_local('faiss-vectorstore',embeddings=embedding,allow_dangerous_deserialization=True)
```

```python
load_db.similarity_search("what is self attention")[0]
```

```python
#convert own query into embedding
question="what is deep learning"
result=db._embed_query(question)
print(result)
```

```python
#convert list of query into embedding

question1=(['what is transformer',
            'what is gen ai'])

result1=db._embed_documents(question1)
print(result1)
```

```python
#similarity search in db
result2=db.similarity_search(question)
result2
```

```python
#similarity search with 
result2=db.similarity_search_with_score(question)
result2
```

```python
#retriever 
vector_db=db.as_retriever()
```

```python
vector_db.invoke("what is transformer")
```
